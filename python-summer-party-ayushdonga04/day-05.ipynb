## Question 2

Using the cleaned data, calculate the total pre-sale orders per month for each region and demographic group.

```python
# Step 1: Remove rows with missing values
cleaned_data = pre_sale_data.dropna()

# Step 2: Convert 'pre_order_date' to datetime format
cleaned_data['pre_order_date'] = pd.to_datetime(cleaned_data['pre_order_date'])

# Step 3: Extract 'Year-Month' from pre_order_date
cleaned_data['order_month'] = cleaned_data['pre_order_date'].dt.to_period('M').astype(str)

# Step 4: Group by 'order_month', 'region', and 'demographic_group' and sum 'pre_order_quantity'
result = cleaned_data.groupby(
    ['order_month', 'region', 'demographic_group'],
    as_index=False
)['pre_order_quantity'].sum()

# Step 5: Rename the column for clarity
result = result.rename(columns={'pre_order_quantity': 'total_pre_order_quantity'})

# Step 6: Print the final result
print(result)
```

## Question 3

Predict the total pre-sales quantity for each region for September 2024. Assume that growth rate from August to September, is the same as the growth rate from July to August in each region.

```python
# Step 1: Clean and prepare the data
cleaned_data = pre_sale_data.dropna()
cleaned_data['pre_order_date'] = pd.to_datetime(cleaned_data['pre_order_date'])

# Step 2: Extract month and year
cleaned_data['month'] = cleaned_data['pre_order_date'].dt.month
cleaned_data['year'] = cleaned_data['pre_order_date'].dt.year

# Step 3: Filter data for July and August 2024
filtered_data = cleaned_data[
    ((cleaned_data['year'] == 2024) & (cleaned_data['month'] == 7)) |
    ((cleaned_data['year'] == 2024) & (cleaned_data['month'] == 8))
]

# Step 4: Group by region and month, and sum pre_order_quantity
monthly_totals = filtered_data.groupby(['region', 'month'])['pre_order_quantity'].sum().unstack()

# Step 5: Calculate growth rate from July (7) to August (8)
monthly_totals['growth_rate'] = (monthly_totals[8] - monthly_totals[7]) / monthly_totals[7]

# Step 6: Predict September (based on August * (1 + growth_rate))
monthly_totals['predicted_september_quantity'] = monthly_totals[8] * (1 + monthly_totals['growth_rate'])

# Step 7: Final output
predictions = monthly_totals[['predicted_september_quantity']].reset_index()
print(predictions)
```
